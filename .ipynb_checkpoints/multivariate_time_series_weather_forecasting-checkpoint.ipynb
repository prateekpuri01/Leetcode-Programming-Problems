```python
weather_files=glob.glob('./*.json')
dfs=[0]*len(weather_files)
for i in range(len(weather_files)):
    dfs[i]=pd.read_json(weather_files[i])
```

```python
weather_df=pd.concat(dfs).drop_duplicates()
attributes_to_keep=['AWND','PRCP','SNOW','TAVG','TMAX','TMIN']
weather_df=weather_df[weather_df['datatype'].apply(lambda x: x in attributes_to_keep)]
weather_df=pd.pivot_table(weather_df, values='value', index=['date'],
                    columns=['datatype'], aggfunc=np.sum).reset_index()
weather_df['date']=pd.to_datetime(weather_df['date'])
weather_df = weather_df.apply(lambda x: x.fillna(x.mean()),axis=0)[500:]
weather_df['date']=weather_df['date'].apply(lambda x: str(x.month)+'-'+str(x.year))
weather_df=weather_df.groupby(by='date').mean().reset_index()
weather_df['date']=pd.to_datetime(weather_df['date'])
weather_df.sort_values('date',inplace=True)
```

```python
import statsmodels.api as sm

def tsplot(y, title, lags=None, figsize=(10, 6)):
    fig = plt.figure(figsize=figsize)
    layout = (2, 2)
    ts_ax = plt.subplot2grid(layout, (0, 0))
    hist_ax = plt.subplot2grid(layout, (0, 1))
    acf_ax = plt.subplot2grid(layout, (1, 0))
    pacf_ax = plt.subplot2grid(layout, (1, 1))

    y.plot(ax=ts_ax)
    ts_ax.set_title(title, fontsize=12, fontweight='bold')
    y.plot(ax=hist_ax, kind='hist', bins=25)
    hist_ax.set_title('Histogram')
    sm.graphics.tsa.plot_acf(y, lags=lags, ax=acf_ax)
    sm.graphics.tsa.plot_pacf(y, lags=lags, ax=pacf_ax)
    sns.despine()
    plt.tight_layout()
    plt.show()
    return None
```

```python
#first thing we should do is make a sarimax model for prediciton
df_length=len(weather_df['TAVG'])
y_train=weather_df.set_index('date')['TAVG'][:int(df_length*.8)]
y_test=weather_df.set_index('date')['TAVG'][int(df_length*.8):]
tsplot(y_train,'Average Temp.',lags=50)
```

```python
y_train_diff=y_train.diff().dropna()
sm.tsa.stattools.adfuller(y_train_diff.dropna())
#seems like differencing of one makes things okay
```

```python
tsplot(y_train_diff,'Average Temp.',lags=12)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

arima_order=[[[(i,j,k) for i in range(3)] for j in range(3)] for k in range(3)]
seasonal_order=[[[(i,j,k,12) for i in range(3)] for j in range(3)] for k in range(3)]
arima_order=sum(sum(arima_order,[]),[])
seasonal_order=sum(sum(seasonal_order,[]),[])
```

```python
import time 
#fit the model
best_aic=float('inf')
for sparams in seasonal_order:
    for params in arima_order:
        try:
            start = time.time()
            model = SARIMAX(y_train, order=params, seasonal_order=sparams)
            model_fit = model.fit()
            if model_fit.aic < best_aic:
                best_aic=model_fit.aic
                best_params=params
                best_sparams=sparams
            end = time.time()
            print('Model Fitting Time:', end - start)
        except:
            continue
```

```python
print("best aic value: {}".format(best_aic))
print("best params: {}".format(best_params))
print("best seasonal params: {}".format(best_sparams))
```

```python
model = SARIMAX(y_train, order=best_params, seasonal_order=best_sparams)
model_fit = model.fit()
```

```python
predictions = model_fit.forecast(len(y_test))
predictions = pd.Series(predictions, index=y_test.index)
residuals = np.array(y_test) - np.array(predictions)
print("MSE: {}".format(np.sum(residuals**2)))
```

```python
plt.figure()
plt.hist(residuals)
```

```python
prediction_band=model_fit.get_prediction(start=25,end=model.nobs+len(y_test))
predict_ci=prediction_band.conf_int(alpha=0.01)

plt.plot(weather_df['date'],weather_df['TAVG'])
plt.plot(predictions)
plt.fill_between(weather_df['date'][24:], predict_ci.iloc[:, 0], predict_ci.iloc[:, 1], alpha=0.15)
plt.xlabel('time')
plt.ylabel('temp')
```

```python
from statsmodels.tsa.arima_model import ARIMA

for params in arima_order:
    try:
        start = time.time()
        model = ARIMA(y_train, order=params)
        model_fit = model.fit()
        if model_fit.aic < best_aic:
            best_aic=model_fit.aic
            best_params=params
        end = time.time()
        print('Model Fitting Time:', end - start)
    except:
        continue
```

```python
model = ARIMA(y_train, order=best_params)
model_fit = model.fit()
```

```python
predictions = model_fit.forecast(len(y_test))[0]
predictions = pd.Series(predictions, index=y_test.index)
residuals = np.array(y_test) - np.array(predictions)
print("MSE: {}".format(np.sum(residuals**2)))
```
