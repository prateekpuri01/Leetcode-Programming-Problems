```python
import random
import numpy as np
from matplotlib import pyplot as plt
```

```python
pop_mean=5
pop_var=5
rand_A=np.array([random.gauss(pop_mean,pop_var) for i in range(10000)])
rand_B=np.array([random.gauss(pop_mean,pop_var) for i in range(10000)])
```

Let's say we are doing a one sample t-test. We want to test if values from rand_A are generated by a process with a mean different that 0. 

the t statistics in this case is 

$t_s = \frac{\bar{X}-\mu}{S/\sqrt{n}}$

here $S$ is the sample standard deviation given as 

$S = \frac{1}{n-1}\sum_{i=0}^n (X_i -\bar{X})^2 $

Here $S$ is the estimate of the population standard deviation obtained from looking at a particular obtained sample.
Why the $(n-1)$ in the denominator?

Now when calculating S, we actually don't know what the true population mean is, as we would need to know to calcualte a standard deviation. All we know is what the *sample mean* is. If we try to estimate the population from the sample mean, the obtained value is always less than or equal to the actual population standard deviation. The 1/(n-1) term helps accounts for this bias 

If you don't believe me, see below

```python
std_devs=[]
sample_std_devs=[]
num_samples=1000


for i in range(10000):
    sub_sample=np.array(random.sample(list(rand_A),num_samples))
    std_devs.append((1/(num_samples)*sum((sub_sample-pop_mean)**2)))
    sample_std_devs.append(1/(num_samples-1)*sum((sub_sample-np.mean(sub_sample))**2))
```

```python
plt.hist(std_devs,bins=20,alpha=.5,label='bessel corrected',density=True)
plt.hist(sample_std_devs,bins=20,alpha=.5,label='sample var',density=True)
plt.axvline(25,linestyle='--',color='red',label='popualtion var')
plt.legend()
plt.xlabel('variance')
plt.xlabel('count')

plt.figure()
plt.scatter(sample_std_devs,std_devs)
plt.plot([i for i in range(int(min(std_devs)),int(max(std_devs)),1)], \
         [i for i in range(int(min(std_devs)),int(max(std_devs)),1)],linestyle='--',color='red')

plt.xlabel('sample mean estimate')
plt.ylabel('population mean estimate')
plt.title('variance calc from sample')
```

Now let's see how this can be used to calculate the t-statistic for hypothesis tests

```python
means=[]
t_stats=[]
num_samples=2

for i in range(10000):
    sub_sample=np.array(random.sample(list(rand_A),num_samples))
    sample_mean=np.mean(sub_sample)
    means.append(sample_mean)
    sample_var=1/(num_samples-1)*np.sum((sub_sample-sample_mean)**2)
    t_stats.append((num_samples)**(1/2)*(sample_mean-pop_mean)/(sample_var**(1/2)))
```

```python
plt.hist([i for i in t_stats if abs(i) <10],bins=40,alpha=.5,density=True,label='t statistic')
plt.hist((np.array(means)-pop_mean)*(num_samples)**(1/2)/5,bins=20,alpha=.5,density=True,label='z score')
plt.legend()
plt.xlim(-5,5)
```

This is important in regression analysis. The coefficient in the model are normally distributed.

This can be seen from the nromal equation solution

$\hat{\beta}=(\mathbf{X}^T \mathbf{X})^{-1}\mathbf{X}^T \mathbf{Y}$

with

$\mathbf{Y} = \mathbf{X}\mathbf{\hat{\beta}}+\mathbf{\epsilon}$

so

$\hat{\beta}=(\mathbf{X}^T \mathbf{X})^{-1}\mathbf{X}^T (\mathbf{X}\mathbf{\hat{\beta}}+\mathbf{\epsilon})$

$=\hat{\mathbf{\beta}}+(\mathbf{X}^T \mathbf{X})^{-1}\mathbf{X}^T \mathbf{\epsilon}$

so 

$\mathbf{\beta}-\hat{\mathbf{\beta}}=(\mathbf{X}^T \mathbf{X})^{-1}\mathbf{X}^T \mathbf{\epsilon}$

so we see our coefficient bias is proportional to the random variable $\epsilon$ which is itself normally distributed, so our bias will be normally distributed as well
